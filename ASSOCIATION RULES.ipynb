# **Association Rule Mining**

```
#  Display the list of transactions

trans = []
for i in range(0,5176):   # all rows
    trans.append([str(df1.values[i,j]) for j in range (0,20)])  # all columns
trans

trans[0]   # 0th column up to 1,2,3,4,5,6,7,8,...19
```

```
from mlxtend.frequent_patterns import apriori,association_rules
from apyori import apriori
rules = apriori(transactions = trans,min_support = 0.003,
                min_confidence = 0.2,
                min_lift = 2,
                min_length = 2,
                max_length = 2)
reports = list(rules)
reports
```

```
# creating empty list
a = []
b = []
c = []
d = []
e = []
for i in range(0,18):
    a.append(reports[i][2][0][0]) # base item
    b.append(reports[i][2][0][1]) # add item
    c.append(reports[i][1]) # support item
    d.append(reports[i][2][0][2]) # confidence item
    e.append(reports[i][2][0][3]) # lift item
```

```
Base = pd.DataFrame(a)
add = pd.DataFrame(b)
support = pd.DataFrame(c)
confidence = pd.DataFrame(d)
lift = pd.DataFrame(e)
```

```
df_new = pd.concat([Base,add,support,confidence,lift],axis=1)
df_new.columns = ["Base","add","support","confidence","lift"]
df_new
```

```
# high chance to lift the item

new = df_new[df_new.lift>3]
new
```

# **Interview Questions**

## **1.	What is lift and why is it important in Association rules?**

lift is a metric that measures the strength of the relationship between two items, taking into account their individual frequencies. It indicates how much more likely the second item is to occur when the first item is present, compared to their independent occurrence. A lift value greater than 1 suggests a positive association, while a value less than 1 indicates a negative association.

## **2.	What is support and Confidence. How do you calculate them?**

support measures how frequently an item or itemset appears in a dataset, while confidence measures the strength of the association between two items

*   Support is calculated by dividing the number of transactions containing the itemset by the total number of transactions.
*   Confidence is calculated by dividing the number of transactions containing both
items by the number of transactions containing the first item

## **3.	What are some limitations or challenges of Association rules mining?**

*   generating a vast number of potentially irrelevant rules
*   requiring careful selection of metrics and thresholds
*   struggling with noisy or incomplete data
*   It can also be computationally expensive for large datasets and may not always lead to easily interpretable results.
